```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성
np.random.seed(42)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)

# 데이터 시각화
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Generated Data and Regression Lines for n_iterations = 1 and 2')

# 학습률 및 초기화
lr = 0.1
w = np.ones((x.shape[1], 1))
b = 0
m = len(x)
for n_iterations in [1, 10]:
    w = np.ones((x.shape[1], 1))
    b = 0
    for iteration in range(n_iterations):
        # 예측값 계산
        z = x.dot(w) + b
        loss = (z - y)

        # 그래디언트 계산
        w_grad = 2/m * x.T.dot(loss)
        b_grad = 2/m * np.sum(loss)
        print(w_grad, b_grad)
        # w와 b 업데이트
        w -= lr * w_grad
        b -= lr * b_grad

    # 결과 출력
    plt.plot(x, x.dot(w) + b, label=f'n_iterations = {n_iterations}')

plt.legend()
plt.show()
```

## Numpy Random

- rand: 0부터 1사이의 균일 분포를 가진 난수를 생성합니다.
- randn: 평균 0, 표준편차 1의 정규 분포를 가진 난수를 생성합니다.
- randint: 주어진 최소, 최대 범위 사이의 정수 난수를 생성합니다.
- choice: 주어진 1차원 배열에서 임의의 샘플을 생성합니다.
- shuffle: 배열의 순서를 무작위로 섞습니다.

```python
# rand: 0과 1 사이의 균일 분포에서 난수 생성
print("1. rand 함수 사용 예:")
rand_array = np.random.rand(5)  # 5개의 난수 생성
print(rand_array)
```

```python
# randn: 표준 정규 분포에서 난수 생성
print("2. randn 함수 사용 예:")
randn_array = np.random.randn(5)  # 5개의 난수 생성
print(randn_array)
```

```python
# randint: 정수 난수 생성
print("3. randint 함수 사용 예:")
randint_array = np.random.randint(0, 10, size=5)  # 0부터 9까지의 정수 중에서 5개의 난수 생성
print(randint_array)
```

```python
# choice: 배열에서 무작위 샘플 추출
print("4. choice 함수 사용 예:")
my_array = [10, 20, 30, 40, 50]
choice_array = np.random.choice(my_array, size=3)  # my_array에서 3개의 원소를 무작위로 선택
print(choice_array)
```

```python
# shuffle: 배열 순서 섞기
print("5. shuffle 함수 사용 예:")
shuffle_array = np.array([1, 2, 3, 4, 5])
np.random.shuffle(shuffle_array)
print(shuffle_array)
```

- choice: 주어진 1차원 배열에서 임의의 샘플을 생성합니다. replace 인자를 사용하면 중복 선택을 허용하거나 방지할 수 있습니다.
- permutation: 입력 배열의 복사본을 무작위로 섞은 후 반환합니다. 만약 입력이 정수 n이면, range(n)의 순열을 반환합니다.


```python
# choice: 배열에서 무작위 샘플 추출
print("1. choice 함수 사용 예:")
my_array = [10, 20, 30, 40, 50]
choice_array = np.random.choice(my_array, size=3, replace=False)  # replace=False로 중복 선택 방지
print(choice_array)

# permutation: 배열의 순열 생성
print("\n2. permutation 함수 사용 예:")
perm_array = np.array([1, 2, 3, 4, 5])
permuted_array = np.random.permutation(perm_array)
print(permuted_array)
```

```python
# 데이터 생성
np.random.seed(5000)
x = 2 * np.random.rand(100, 1)
y = (x > 1).astype(int)  # 1 if x > 1, else 0

# 시그모이드 함수 정의
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# 데이터 시각화
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Generated Data and Logistic Regression')

# 학습률 및 초기화
lr = 0.1
w = np.ones((x.shape[1], 1))
b = 0
m = len(x)

# 학습 및 결과 출력
for n_iterations in [1, 5000]:
    w = np.ones((x.shape[1], 1))
    b = 0
    for iteration in range(n_iterations):
        # 예측값 계산
        z = x.dot(w) + b
        y_pred = sigmoid(z)


        # 그래디언트 계산
        loss = y_pred - y
        w_grad = 2/m * x.T.dot(loss)
        b_grad = 2/m * np.sum(loss)
        print(w_grad, b_grad)
        # w와 b 업데이트
        w -= lr * w_grad
        b -= lr * b_grad

    # 결과 출력
    plt.plot(x, sigmoid(x.dot(w) + b), label=f'n_iterations = {n_iterations}')

plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

model_sk = LogisticRegression().fit(x, y)

xx = np.linspace(-3, 3, 100)
mu = 1.0/(1 + np.exp(-model_sk.coef_[0][0]*xx - model_sk.intercept_[0]))
plt.plot(xx, mu)
plt.scatter(x, y, c=y, s=100, edgecolor="k", lw=2)
plt.scatter(x, model_sk.predict(x), label=r"$\hat{y}$", marker='s', c=y,
            s=100, edgecolor="k", lw=1, alpha=0.5)
plt.xlim(-3, 3)
plt.xlabel("x")
plt.ylabel(r"$\mu$")
plt.title(r"$\hat{y}$ = sign $\mu(x)$")
plt.legend()
plt.show()
```
