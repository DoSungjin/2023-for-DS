### 2.1 함수

####[2.1.1] 
우리는 지난 시간에 print(), type(), float(), int(), str(), 등과 같이 특정 작업을 수행하는 여러 가지 명령어들을 배웠습니다. 이러한 명령어들의 정확한 명칭 '함수'입니다. 일반적으로 수학에서 말하는 함수는 어떤 입력값을 받으면 그에 대응하는 출력값을 반환하는 것을 의미합니다. 파이썬의 함수 또한 끝에 항상 소괄호 ()를 달고 다니는데 그 소괄호 사이에 '인자 (argument)'라고 부르는 입력값을 적을 수 있습니다. 

```
def 함수명(입력값):
    
    <수행할 작업>

    return 결과값
```

````python
# *****[2.1.1] 아래의 코드를 실행하세요*****
def square_sum(x, y):
    x_square = x ** 2
    y_square = y ** 2
    result = x_square + y_square
    return result

print(square_sum(1, 2))
print(square_sum(2, 3))
````

###2.2 NumPy 라이브러리의 활용

####[2.2.1] 
리스트는 크기가 가변적이고 다양한 자료형을 담을 수 있는 유연함이 있지만, 반복된 여러 연산을 처리하는 데에 느리다는 단점이 있습니다. 이러한 리스트의 단점을 극복하고 수치연산에서 많이 쓰이는 자료형이 바로 NumPy에 구현된 배열(array)입니다. 한 자료형이 담긴 리스트에 np.array() 함수를 사용하면 배열로 바꾸어 입력할 수 있습니다. 그 밖에도 특정 크기의 배열에 0을 가득 채우는 np.zeros(), 특정 범위 내 정해진 갯수의 숫자 배열을 만드는 np.linspace(), 특정 범위를 일정한 크기로 나눈 숫자 배열을 만드는 np.arange() 등이 자주 활용됩니다. 

````python
# *****[2.2.1] 아래의 코드를 실행하세요*****
data_list = [1, 2, 3, 4, 5]
data_array = np.array(data_list)
print(type(data_list))
print(type(data_array))

array_zeros = np.zeros(10)
print(array_zeros)

array_linspace = np.linspace(10, 100, 4)
print(array_linspace)

array_arange = np.arange(10, 100, 4)
print(array_arange)
````
